**Bonus Problem:** route length **R**, **N** cells.

```python
def count_routes(ncell=3, nstep=2):
    cell = [1] + [0] * (ncell - 1)
    for istep in range(nstep):
        print(istep, cell)
        updated = list(cell)
        for i in range(len(cell)):
            if i > 0:
                updated[i-1] += cell[i]
            if i < ncell-1:
                updated[i+1] += cell[i]
        cell = updated
    print(nstep, cell)
```

```python
count_routes()
```

```python
count_routes(4,5)
```

```python
count_routes(7, 35)
```
